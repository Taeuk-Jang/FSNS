```python
import numpy as np 
import pickle

import matplotlib.pyplot as plt
```

```python
dataname = 'german'
```

```python
fr = pickle.load(open(f'results/{dataname}/robust_fr_dict.pkl', 'rb'))
base = pickle.load(open(f'results/{dataname}/robust_baseline_dict.pkl', 'rb'))
fsns = pickle.load(open(f'results/{dataname}/robust_FSNS_dict.pkl', 'rb'))
```

```python
#German
plt.figure(figsize = (8,5))

plt.plot([fr[key]['EqOdds'] for key in [0.1* i for i in [1,2,3,4]]], [fr[key]['Acc'] for key in [0.1* i for i in [1,2,3,4]]], label = 'FR-Train')
plt.plot([np.array([i.item() for i in base[key]['EqOdds']]).mean() for key in [0.1* i for i in range(1,6)]], \
             [np.array([i.item() for i in base[key]['Acc']]).mean() for key in [0.1* i for i in range(1,6)]], label = 'Baseline')
plt.plot([fsns[key]['EqOdds'] for key in [0.1* i for i in [4,2,1,5]]], [fsns[key]['Acc'] for key in [0.1* i for i in [4,2,1,5]]], label = 'FSNS')
# plt.xlim(0.05, 0.45)

plt.ylabel("Acc", fontsize = 14)
plt.xlabel("EqOdds", fontsize = 14)
plt.legend(fontsize = 13)
plt.tight_layout()
plt.savefig('./plot_fig/{}/robust.pdf'.format(dataname), dpi = 400)
```
